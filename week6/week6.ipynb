```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import sympy as sm
import scipy as sc
```

Differential equation

$y' = 5e^{-x} -2y$

```python
#Analytical solution
from sympy.abc import x

y = sm.Function('y') #define that y is a function
#Write the equation in to the form where we have ######only ZERO at the other side#################
#y' -5*e^(-x) + 2y

#Write the terms in a different format
# y' --> sm.Derivative(y(x),x)
# -5*e^(-x) --> 5*sm.exp(-x)
# 2y --> 2*y(x)


result = sm.dsolve(sm.Derivative(y(x),x) - 5*sm.exp(-x) + 2*y(x), y(x))

result
```

```python
#Numerical solution

#Define the differential equation as a Python function
#Must be in the form y' = 5 * e^(-x) - 2y

def diffun(y,x):
    dydx = 5*np.exp(-x) - 2*y
    return dydx

#Define value of x ( rqange where we define the solution)
x = np.linspace(0,10,101)
                
#Boundary condition
y0 = 1 #y(0) = 1

y = sc.integrate.odeint(diffun,y0,x)

#From analytical solution, using boundary condition y(0) = 1
#--> (C1 + 5)=1 ---> C1 = -4

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')

#add analytical solution to the plot
C= -4
analytical = (C*np.exp(-x)+5)*np.exp(-x)
plt.plot(x, analytical, 'r--')
plt.legend(['Numerical solution', 'Analytical solution'])
```
