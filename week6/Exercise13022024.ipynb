```python
import numpy as np
import matplotlib.pyplot as plt

#import for analytical solution(Solove equation and get equation,pen and paper style)
#Symbolic Physon
import sympy as sm

#Numerical solution (Define the equation in terms of numbers and get the result as numbers )
#Symbolic Physon
import scipy as sc 
```

```python
#Analytical solution ,using scipy.dsolve function

#dsolve requires that we write the equation in format where we have only zero on other side
#Now we already haave this format

#We must write each term in a correct way

#y'--> sm.Derivative(y(x),x) #Compare to dy(x)/dx   
#7y --> 7*y(x)
from sympy.abc import x #x is a variable
y = sm.Function('y') #y is a function

solution = sm.dsolve(sm.Derivative(y(x),x) + 7*y(x),y(x)) 
#solution = sm.dsolve(sm.Derivative(y'+7,y) 

#Inside dsolve define the differential equation and what are we solving for
solution
```

```python
#Numerical solution
#This is important because many differential equations don't have analytical solution	

#Let's write the equation in the different format
#y' - 7y = 0	--> dydx +7*y  = 0 --> dydx = -7*y

#Write the differential equation as a Python function
#Inputs are y and x , returns y'
def f11a(y,x):
    dydx = -7*y #Same as y' +7y = 0
    return dydx

#Difine the range where we look for a solution
X = np.linspace(0,10,101) #X is an array of values from 0 to 10 between 0.1

#Boundary condition (numerical solution cannot include C)
y0 = 2 # Defines that the value of y is 2 in the first element of X --> y(0) = 2

num_solution = sc.integrate.odeint(f11a,y0,X) #Equation, boundary condition, X values
num_solution

#Is this the curve y(x) = C*e^(-7x) with some value of C? Let's investigate:
#Boundary condition y(0) = 2 --> y(0) = C --> C = 2
#--> y(x) = 2*e^(-7x)
plt.figure(figsize=(7,5))
plt.plot(X,num_solution,'*') #Numerical solution

C = 2

plt.plot(X,2*np.exp(-7*X),'r--') #Analytical solution

plt.legend(['Numerical solution','Analytical solution'])

plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.show()
```

y' + cos(x)y = 0

```python
#Analytical solution
#y' + cos(x)y = 0

#Terms in the correct format
#y' --> sm.Derivative(y(x),x)
#cos(x)y --> sm.cos(x)*y(x)
from sympy.abc import x #x is a variable
y = sm.Function('y') #y is a function

solution = sm.dsolve(sm.Derivative(y(x),x) + sm.cos(x)*y(x),y(x))
solution
```

```python
#Numerical solution

#Define the function
#y' + cos(x)y = 0 --> dydx = -np.cos(x)*y

def f11b(y,x):
    dydx = -np.cos(x)*y
    return dydx

#Difine the range where we look for a solution
X = np.linspace(0,10,101) #X is an array of values from 0 to 10 between 0.1

#Boundary condition (numerical solution cannot include C)
y0 = 2 # Defines that the value of y is 2 in the first element of X --> y(0) = 2

num_solution = sc.integrate.odeint(f11b,y0,X) #Equation, boundary condition, X values
num_solution

plt.figure(figsize=(7,5))
plt.plot(X,num_solution,'*') #Numerical solutionkorewo


```

```python
#Analytical solution
#xy' + 3*x^2*y = 0    

#Let's write the terms in the proper format
#xy' --> x*sm.Derivative(y(x),x) #Compare to x*dy(x)/dx
#3*x^2*y --> 3*x**2*y(x)

from sympy.abc import x  # x
y = sm.Function('y')     # y

solution = sm.dsolve(x*sm.Derivative(y(x),x) + 3*x**2*y(x), y(x))
solution
```

```python

# Define the function for the differential equation
# xy' + 3*x^2*y = 0 
#--> y' +3xy = 0 --> y' = -3yx
#dydx = -3*y*x


def f11c(y, x):
    dydx = -3*y*x
    return dydx


###################### copy-pasete EXCEPT for the function name#####################

# Define the range of x values
X = np.linspace(0.1, 10, 101) # x

# Boundary condition (initial value)
y0 = 2 # y(0.1) = 2


num_solution = sc.integrate.odeint(f11c,y0,X) #Equation, boundary condition, X values
num_solution

plt.figure(figsize=(7,5))
plt.plot(X,num_solution,'*') #Numerical solution
C = 2
plt.plot(X,C*np.exp(-3/2*X**2),'r--') #Analytical solution
plt.legend(['Numerical solution','Analytical solution'])
```
