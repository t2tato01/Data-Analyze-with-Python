```python
#To access only single column:
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt#df['Time(s)']
```

```python

#Let´s make a symple graphics
plt.figure(figsize=(15,8))
plt.subplot(3,1,1)
plt.plot(df['Time (s)'], df['Linear Acceleration x (m/s^2)'])
plt.grid()
plt.title('Acceleration x')
plt.xlabel('Time (s)')
plt.ylabel('a_x')
plt.axis([0, 10, -30, 30])


plt.subplot(3,1,2)
plt.plot(df['Time (s)'], df['Linear Acceleration y (m/s^2)'])
plt.grid()
plt.title('Acceleration y')
plt.xlabel('Time (s)')
plt.ylabel('a_y')
plt.axis([0, 10, -30, 30])

plt.subplot(3,1,3)
plt.plot(df['Time (s)'], df['Linear Acceleration z (m/s^2)'])
plt.grid()
plt.title('Acceleration z')
plt.xlabel('Time (s)')
plt.ylabel('a_z')
plt.axis([0, 10, -30, 30])
plt.show()


```
